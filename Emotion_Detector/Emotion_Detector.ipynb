```python
import os
import cv2
import numpy as np
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255,
                                   shear_range=0.2,
                                   zoom_range=0.2,
                                   horizontal_flip=True,
                                   validation_split=0.2)  

training_set = train_datagen.flow_from_directory('/kaggle/input/emotion-recognition-dataset/dataset',
                                                 target_size=(64, 64),
                                                 batch_size=32,
                                                 class_mode='categorical',
                                                 subset='training')

validation_set = train_datagen.flow_from_directory('/kaggle/input/emotion-recognition-dataset/dataset',
                                                   target_size=(64, 64),
                                                   batch_size=32,
                                                   class_mode='categorical',   
                                                   subset='validation') 
```

```python
cnn = tensorflow.keras.models.Sequential()
```

```python
cnn.add(tensorflow.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
```

```python
cnn.add(tensorflow.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tensorflow.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tensorflow.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tensorflow.keras.layers.Conv2D(filters=128, kernel_size=3, activation='relu'))
cnn.add(tensorflow.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tensorflow.keras.layers.Flatten())
```

```python
cnn.add(tensorflow.keras.layers.Dense(units=128, activation='relu'))
```

```python
cnn.add(tensorflow.keras.layers.Dense(units=6, activation='softmax'))
```

```python
cnn.compile(optimizer='adam',
            loss='categorical_crossentropy',  # Use categorical crossentropy for multi-class classification
            metrics=['accuracy'])
```

```python
cnn.summary()
```

```python
cnn.fit(x=training_set,
        validation_data=validation_set,
        epochs=25,
        steps_per_epoch=len(training_set),
        validation_steps=len(validation_set))
```

```python
cnn.fit(x = training_set, validation_data = validation_set, epochs = 25)
```

```python
cnn.fit(x = training_set, validation_data = validation_set, epochs = 15)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

path='/kaggle/input/emotion-recognition-dataset/dataset/Sad/005f95668c97114715e3ea793f91b358bebe83d15b7fb62ec08063af.jpg'

# Load and preprocess the test image
test_image = image.load_img(path, target_size=(64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis=0)

# Make a prediction
result = cnn.predict(test_image)

# Assuming you have the class indices mapping
class_indices = training_set.class_indices
# Reverse the mapping to get class names
class_names = {v: k for k, v in class_indices.items()}

# Get the index of the class with the highest probability
predicted_class_index = np.argmax(result[0])

# Get the class name
prediction = class_names[predicted_class_index]

print(f'Prediction: {prediction}')
```
